```
from gtts import gTTS
import os
import librosa, librosa.display
import numpy, scipy, matplotlib.pyplot as plt, IPython.display as ipd, sklearn

mytext = "Hello. Welcome to the College of Engineering and Computer Science at the University of Tennessee."
mytext = mytext + "My name is Hong Qin. Hong is the first word of Hong Kong, and it means great. Qin is pronounced as c h i n."
mytext = mytext + " So, I have a great chin. I am a faculty in Computer Science. I teach data science and software development."
mytext = mytext + "My research field is computational biology. I research how we can live longer and healthier lives."
mytext = mytext + "My students and I are using graph based deep learning  and transfer learning to identify the conserved lifespan extension mechasnims. "
mytext = mytext + "I recently research how to predict pandemics, how to predict the pathogencity of new variants."
mytext = mytext + " I  research how to detect fake images, how to build robust deep learning methods."
mytext = mytext + " I reserch how to build reliable and resilent engineered systems. "
mytext = mytext + " I will give you a few examples to show how computer science can be useful"

final_file = gTTS(text = mytext, lang='en')
final_file.save("test.mp3")
```
