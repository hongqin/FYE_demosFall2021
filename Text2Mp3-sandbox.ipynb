```
from gtts import gTTS
import os
import librosa, librosa.display
import numpy, scipy, matplotlib.pyplot as plt, IPython.display as ipd, sklearn

string_of_text = 'Hello. Welcome to UTC. Chattanooga is a great place.'
final_file = gTTS(text = string_of_text, lang='en')
final_file.save("test.mp3")
```

```
x, sr = librosa.load("test.mp3")
ipd.Audio(x, rate=sr)
```
